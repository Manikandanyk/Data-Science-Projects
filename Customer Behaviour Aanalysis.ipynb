```python
import pandas as pd
```

```python
sql = """SELECT 
    stage, 
    COUNT(DISTINCT CustomerID) AS customers_count
FROM customer_journey
GROUP BY stage
ORDER BY customers_count 
"""
```

```python
curr = db_connection.cursor()
```

```python
curr.execute(sql)
```

```python
CJA = curr.fetchall()
```

```python
curr.column_names
```

```python
CJA = pd.DataFrame(CJA, columns = curr.column_names)
```

```python
CJA
```

```python
sql = """SELECT 
    stage, 
    AVG(Duration) AS avg_time_spent
FROM customer_journey
GROUP BY stage;"""
```

```python
curr = db_connection.cursor()
```

```python
curr.execute(sql)
```

```python
Avg_Duration = curr.fetchall()
```

```python
curr.column_names
```

```python
Avg_Duration = pd.DataFrame(Avg_Duration, columns = curr.column_names)
```

```python
Avg_Duration
```

```python
sql = """SELECT 
    ProductID, 
    COUNT(*) AS total_reviews,
    ROUND(AVG(rating),2) AS avg_rating
FROM customer_reviews
GROUP BY ProductID
ORDER BY avg_rating DESC;"""
```

```python
curr = db_connection.cursor()
```

```python
curr.execute(sql)
```

```python
Avg_Rating = curr.fetchall()
```

```python
curr.column_names
```

```python
Avg_Rating = pd.DataFrame(Avg_Rating, columns = curr.column_names)
```
